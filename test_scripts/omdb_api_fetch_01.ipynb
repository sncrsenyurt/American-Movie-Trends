### Movie Success Factors (Placeholder Title)

```python
# Dependencies
import pandas as pd
import requests
import json
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt


#API Key
from api_keys import omdb_api_key
```

```python
#Read csv

movie_data = []

movies = pd.read_csv("resources/AllMovies_2018-2022.csv")

movie_list = [movie for movie in movies["Title"]]

#Go to OMDB and get data

for movie in movie_list:
    try:
        print(f"Fetching data for | {movie}")
        
        url = f"http://www.omdbapi.com/?t={movie}&apikey={omdb_api_key}"

        movie_json = requests.get(url).json()

        movie_data.append({"Title":movie_json["Title"],
                           "Year":movie_json["Year"],
                           "Released":movie_json["Released"],
                           "Runtime":movie_json["Runtime"],
                           "Genre":movie_json["Genre"],
                           "Awards":movie_json["Awards"],
                           "Metascore":movie_json["Metascore"],
                           "imdbRating":movie_json["imdbRating"],
                           "imdbVotes":movie_json["imdbVotes"],
                           "imdbID":movie_json["imdbID"],
                           "BoxOffice":movie_json["BoxOffice"],
                           "Language":movie_json["Language"]})
        print("Success")
        
    except:
        print("Movie not found...")
        pass

```

```python
movie_df = pd.DataFrame(movie_data)
movie_df.head()
```

```python
movie_df = movie_df.to_csv("resources/all_movies.csv",index=False)
```
