```python
# Dependencies
import pandas as pd
import requests
import json
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
tmdb_df = pd.read_csv("../moviedata/tmdb_filtered_by_date.csv")
tmdb_df.head()
```

```python
tmdb_budget_df = tmdb_df[["title","budget"]]
tmdb_budget_df.head()
```

```python
tmdb_budget_max = tmdb_budget_df.groupby("title")["budget"].max()
```

```python
tmdb_budget_max
```

```python
tmdb_budget_min = tmdb_budget_df.groupby("title")["budget"].min()
tmdb_budget_min
```

```python
tmdb_budget_max.index
```

```python
tmdb_budget_df_2 = pd.DataFrame({"budget_max":tmdb_budget_max,
                                 "budget_min":tmdb_budget_min})
tmdb_budget_df_3 = tmdb_budget_df_2.reset_index()
tmdb_budget_df_3
```

```python
clean_df = pd.read_csv("resources/all_movies.csv")
clean_df
```

```python
tmdb_budget_df_4 = tmdb_budget_df_3.rename(columns={"title": "Title"})
tmdb_budget_df_4.head()
```

```python
merged_df = pd.merge(clean_df,tmdb_budget_df_4,how="left",on="Title")
```
